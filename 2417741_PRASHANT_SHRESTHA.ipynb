# TASK **3**

```python
# Finding outlier countries in South Asia based on Score and GDP per capita

# Calculating IQR for GDP
gdpQ1 = south_asian_DataFrame['Log GDP per capita'].quantile(0.25)
gdpQ3 = south_asian_DataFrame['Log GDP per capita'].quantile(0.75)
gdpIQR = gdpQ3 - gdpQ1

# Defining thresholds for low and high GDP
GdpLow = gdpQ1 - 1.5 * gdpIQR
GdpHigh = gdpQ3 + 1.5 * gdpIQR

# Calculating IQR for Score
scoreQ1 = south_asian_DataFrame['score'].quantile(0.25)
scoreQ3 = south_asian_DataFrame['score'].quantile(0.75)
scoreIQR = scoreQ3 - scoreQ1

# Defining thresholds for low and high scores
scoreLowGdp = scoreQ1 - 1.5 * scoreIQR
scoreHighGdp = scoreQ3 + 1.5 * scoreIQR

# Marking outliers based on the thresholds for GDP and Score
south_asian_DataFrame['Outliers'] = (
    (south_asian_DataFrame['Log GDP per capita'] < GdpLow) |
    (south_asian_DataFrame['Log GDP per capita'] > GdpHigh) |
    (south_asian_DataFrame['score'] < scoreLowGdp) |
    (south_asian_DataFrame['score'] > scoreHighGdp)
)

# Plotting a scatterplot to visualize the outliers
plt.figure(figsize=(10, 6))
plt.scatter(
    south_asian_DataFrame['Log GDP per capita'],
    south_asian_DataFrame['score'],
    c=south_asian_DataFrame['Outliers'].map({True: 'red', False: 'blue'})
)
plt.xlabel('Log GDP per capita')
plt.ylabel('Score')
plt.title('Outliers Detection')
plt.legend(['Non-Outliers', 'Outliers'], loc='best')
```

The scatter plot highlights South Asian countries that stand out based on their Score and Log GDP per capita, with outliers marked in red. These outliers show countries that are either much wealthier or less happy than the rest of the region. For instance, some countries with low GDP but decent happiness scores could indicate strong social support. On the other hand, wealthier countries with lower happiness scores might face different challenges. Analyzing these outliers helps us better understand what affects both GDP and happiness in the region.

**Task 6**

```python
# Loop through each region and its corresponding DataFrame
for region, mixed_df in [('South Asia', south_asian_DataFrame), ('Middle East', middle_DataFrame)]:
    # Get the 25th and 75th percentiles for Score
    ScoreQ1 = mixed_df['score'].quantile(0.25)
    ScoreQ3 = mixed_df['score'].quantile(0.75)
    # Calculate the Interquartile Range (IQR) for Score
    ScoreIQR = ScoreQ3 - ScoreQ1

    # Get the 25th and 75th percentiles for GDP
    gdpQ1 = mixed_df['Log GDP per capita'].quantile(0.25)
    gdpQ3 = mixed_df['Log GDP per capita'].quantile(0.75)
    # Calculate the IQR for GDP
    gdpIQR = gdpQ3 - gdpQ1

    # Set the upper and lower bounds to detect outliers
    scoreLowerBound = ScoreQ1 - 1.5 * ScoreIQR
    scoreUpperBound = ScoreQ3 + 1.5 * ScoreIQR
    gdpLowerBound = gdpQ1 - 1.5 * gdpIQR
    gdpUpperBound = gdpQ3 + 1.5 * gdpIQR

    # Mark rows as outliers if Score or GDP fall outside the bounds
    mixed_df['Outliers'] = (
        (mixed_df['score'] < scoreLowerBound) | (mixed_df['score'] > scoreUpperBound) |
        (mixed_df['Log GDP per capita'] < gdpLowerBound) | (mixed_df['Log GDP per capita'] > gdpUpperBound)
    )

    # Create a scatter plot to show the outliers
    plt.figure(figsize=(10, 6))
    plt.scatter(mixed_df['Log GDP per capita'], mixed_df['score'],
                c=mixed_df['Outliers'].map({True: 'red', False: 'blue'}))
    plt.xlabel('Log GDP per capita')
    plt.ylabel('Score')
    plt.title(f'Outliers Detection for {region}')
    plt.legend(['Outliers (red)', 'Normal data (blue)'])
    plt.show()
```

High Score, Low GDP: Countries that have a high happiness score but low GDP might show that even with low income, people can still be really happy. This could be because of strong cultural values, close-knit communities, or social factors that help improve overall happiness, even if the economy isn't as strong.

Low Score, High GDP: Countries with high GDP but low happiness scores might suggest that while they have more money and a strong economy, the people aren’t necessarily happy. This could be due to issues like inequality, political problems, or social unrest that affect people's well-being despite the country's wealth.

Extreme Values: Countries with very high or low GDP and similarly extreme happiness scores could be influenced by unique political or social situations. These outliers might reflect unusual conditions, like specific government policies or social circumstances, that have a big impact on both the economy and happiness levels.
